```python
import pandas as pd
df = pd.read_csv("fb.csv",parse_dates=['Date'],index_col='Date')
df
```

```python
df.shift(1)
```

```python
df.shift(-1)
```

```python
df['Prev Day Price'] = df['Price'].shift(1)
df
```

```python
df['one Day change'] = df['Prev Day Price'] - df['Price']
df
```

```python
df['5 Day % change'] = (df['Prev Day Price'] - df['Price'].shift(5))*100/df['Price']
df
```
